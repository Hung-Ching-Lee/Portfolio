# 抓菜單
# step1. 辨識該文字是否為菜色
# step2. 若是,則紀錄其菜色名與其右邊水平的數字(價格)

```python
"""
換kernel
"""
import torch
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from transformers import BertTokenizer, BertForSequenceClassification, BertConfig, AdamW

import csv
import json
import numpy as np
import pandas as pd

EPOCHS = 1  # 训练的轮数
BATCH_SIZE = 4  # 批大小
MAX_LEN = 10  # 文本最大长度
max_len = 10 
LR = 1e-5  # 学习率
WARMUP_STEPS = 100  # 热身步骤
T_TOTAL = 1000  # 总步骤
```

```python
# 載入訓練好的模型
config = BertConfig.from_pretrained('bert-base-chinese')
config.num_labels = 2
model = BertForSequenceClassification(config = config)
model.load_state_dict(torch.load('/home/jacklee/ocr/food_classification/bert_cla.ckpt'))
tokenizer = BertTokenizer.from_pretrained('bert-base-chinese')
```

```python
"""
文字轉token
"""
def get_text_token(text):
    text = text.replace(' ','')
    num = MAX_LEN - len(text)
    if num < 0:
        text = text[:max_len]
        text = tokenizer.encode(text) 
        num_temp = max_len - len(text)
        if num_temp > 0:
            for _ in range(num_temp):
                text.append(0)
        # 在开头和结尾加[CLS] [SEP]
        text = [101] + text + [102]
        text = str(text)

    for _ in range(num):
        text = text + '[PAD]'
    text = tokenizer.encode(text)
    num_temp = MAX_LEN - len(text)
    if num_temp > 0:
        for _ in range(num_temp):
            text.append(0)
    text = [101] + text + [102]
    text_token = str(text)
    return text_token 
    
```

```python
# 預測是否為菜色
def predict(text):
    text_tensor = torch.tensor(json.loads(get_text_token(text))).unsqueeze(0)
    pre = model(text_tensor, labels=None)[0].argmax(dim=1).item()
    return pre
```

```python
from PIL import Image
"""
82,
"""
file_name = r'/home/jacklee/ocr/image_data5/IMG_0798.HEIC.jpg'
text = reader.readtext(file_name, detail = 1)
img = Image.open(file_name)
img.thumbnail((800,800), Image.ANTIALIAS)
img
```

```python
#判斷是否為菜色,若是,則記起來
import re
menu_list = []
menu_match = []
for t in text:
    tup = []
    if predict(t[1])==1: 
#         menu_match[0].append(t[0])
        tup.append(t[0])
        tup.append(t[1])
        menu_match.append(tup)
        menu_list.append(t[1])
```

```python
# Blue color in BGR 
import numpy as np
import matplotlib.pyplot as plt
import cv2
plt.figure(figsize=(15, 25)) 
img = cv2.imread(file_name)
color = (255, 0, 0) 
thickness = 6
image1 = cv2.rectangle(img, tuple(menu_match[0][0][0]), tuple(menu_match[0][0][2]), color, thickness)
image2 = cv2.rectangle(img, tuple(menu_match[1][0][0]), tuple(menu_match[1][0][2]), color, thickness)
plt.figure(figsize=(8, 15)) 
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.imshow(cv2.cvtColor(image2, cv2.COLOR_BGR2RGB))
print('菜色是',menu_list)
```

```python
#將被判斷為菜色的文字右邊的資訊拉出來
#條件: 1.與菜色同水平 2.在菜色右邊
total_menu_price = []
for m in menu_match:
    menu_price =[]
    price_list = []
    for t in text:
        avg_high = (int(t[0][1][1])+int(t[0][3][1]))/2
        if (m[0][0][1]<avg_high<m[0][3][1]):
            price_list.append(t)
    #選出裡面最右邊的那一格right_price
    right_price = price_list[0]
    for p in price_list:
        if p[0][1][0]>right_price[0][1][0]:
            right_price = list(p)
            #把屬於該菜色m的的right_price,append起來
            menu_price.append(m)
            menu_price.append(right_price)
            total_menu_price.append(menu_price)
            
```
