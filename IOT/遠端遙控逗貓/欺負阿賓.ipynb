```python
        
import tkinter as tk 
import RPi.GPIO as GPIO
from time import sleep
from steering import Steering
import time

GPIO21 = 18
GPIO20 = 23
振動他pin = 24
餵魷魚絲_pin = 25
雷射_pin = 19 

GPIO.setmode(GPIO.BCM)

GPIO.setup(GPIO21, GPIO.OUT)
GPIO.setup(GPIO20, GPIO.OUT)
GPIO.setup(振動他pin, GPIO.OUT)
GPIO.setup(餵魷魚絲_pin, GPIO.OUT)
GPIO.setup(雷射_pin, GPIO.OUT)



master = tk.Tk()
master.title("欺負阿賓")
master.geometry("950x1000")

GPIO21_state = True
餵魷魚絲_State = True
GPIO20_State = True
振動他_State = True
雷射_State =  True

def 閃他():
	global GPIO21_state
	if GPIO21_state == True:
		GPIO.output(GPIO21, GPIO21_state)
		GPIO21_state = False
		閃阿賓label = tk.Label(master, text="Turned ON", fg="green")
		閃阿賓label.grid(row=1, column=0)
	else:
		GPIO.output(GPIO21, GPIO21_state)
		GPIO21_state = True
		閃阿賓label = tk.Label(master, text="Turned OFF", fg="red")
		閃阿賓label.grid(row=1, column=0)


def 吵他():
	global GPIO20_State
	if GPIO20_State == True:
		GPIO.output(GPIO20, GPIO20_State)
		GPIO20_State = False
		吵阿賓label = tk.Label(master, text="Turned ON", fg="green")
		吵阿賓label.grid(row=1, column=1)
	else:
		GPIO.output(GPIO20, GPIO20_State)
		GPIO20_State = True
		吵阿賓label = tk.Label(master, text="Turned OFF", fg="red")
		吵阿賓label.grid(row=1, column=1)

        
def 振動他():
	global 振動他_State
	if 振動他_State == True:
		GPIO.output(振動他pin, 振動他_State)
		振動他_State = False
		振阿賓label = tk.Label(master, text="Turned ON", fg="green")
		振阿賓label.grid(row=1, column=2)
	else:
		GPIO.output(振動他pin, 振動他_State)
		振動他_State = True
		振阿賓label = tk.Label(master, text="Turned OFF", fg="red")
		振阿賓label.grid(row=1, column=2)
def 餵魷魚絲():

    PWM_FREQ = 50
    STEP=150
    pwm = GPIO.PWM(餵魷魚絲_pin, PWM_FREQ)
    pwm.start(100)
    pwm.ChangeDutyCycle(1)
    time.sleep(1)
    pwm.stop()
def 雷射():
	global 雷射_State
	if 雷射_State == True:
		GPIO.output(雷射_pin, 雷射_State)
		雷射_State = False
		雷射阿賓label = tk.Label(master, text="Turned ON", fg="green")
		雷射阿賓label.grid(row=7, column=3)
	else:
		GPIO.output(雷射_pin, 雷射_State)
		雷射_State = True
		雷射阿賓label = tk.Label(master, text="Turned OFF", fg="red")
		雷射阿賓label.grid(row=7, column=3)
    
'''
相機雲台
'''
cam水平 = 90
cam垂直 = 90


def cam水平轉():
    steering1 = Steering(21, 0, 270, 0, 0, 140, cam水平, cam垂直)
    steering1.setup()
    steering1.specify(cam水平,cam垂直)
    steering1.cleanup()
def cam垂直轉():
    steering1 = Steering(0, 0, 270, 20, 0, 140, cam水平, cam垂直)
    steering1.setup()
    steering1.specify(cam水平,cam垂直)
    steering1.cleanup()
def cam右():
    global cam水平
    if (cam水平>=170):
        cam水平 = 170
    elif (cam水平<=10):
        cam水平 = 10
    cam水平-=10
        

def cam左():
    global cam水平
    if (cam水平>=170):
        cam水平 = 170
    elif (cam水平<=10):
        cam水平 = 10

    cam水平+=10


def cam上():
    global cam垂直
    if (cam垂直>=120):
        cam垂直 = 120
    elif (cam垂直<=50):
        cam垂直 = 50
    cam垂直-=10

def cam下():
    global cam垂直
    if (cam垂直>=120):
        cam垂直 = 120
    elif (cam垂直<=50):
        cam垂直 = 50
    cam垂直+=10    

def cam右轉():
    cam右()
    time.sleep(0.2)
    cam水平轉()
def cam左轉():
    cam左()
    time.sleep(0.2)
    cam水平轉()

def cam上轉():
    cam上()
    time.sleep(0.2)
    cam垂直轉()
def cam下轉():
    cam下()
    time.sleep(0.2)
    cam垂直轉()

    
'''
雷射雲台
'''
雷射水平 = 90
雷射垂直 = 90


def 雷射水平轉():
    steering2 = Steering(12, 0, 270, 0, 0, 140, 雷射水平, 雷射垂直)
    steering2.setup()
    steering2.specify(雷射水平,雷射垂直)
    steering2.cleanup()
def 雷射垂直轉():
    steering2 = Steering(0, 0, 270, 16, 0, 140, 雷射水平, 雷射垂直)
    steering2.setup()
    steering2.specify(雷射水平,雷射垂直)
    steering2.cleanup()
def 雷射右():
    global 雷射水平
    if (雷射水平>=130):
        雷射水平 = 130
    elif (雷射水平<=10):
        雷射水平 = 10
    雷射水平-=10
        

def 雷射左():
    global 雷射水平
    if (雷射水平>=130):
        雷射水平 = 130
    elif (雷射水平<=10):
        雷射水平 = 10

    雷射水平+=10


def 雷射上():
    global 雷射垂直
    if (雷射垂直>=130):
        雷射垂直 = 130
    elif (雷射垂直<=60):
        雷射垂直 = 60
    雷射垂直-=10

def 雷射下():
    global 雷射垂直
    if (雷射垂直>=130):
        雷射垂直 = 130
    elif (雷射垂直<=50):
        雷射垂直 = 50
    雷射垂直+=10    

def 雷射右轉():
    雷射右()
    time.sleep(0.2)
    雷射水平轉()

def 雷射左轉():
    雷射左()
    time.sleep(0.2)
    雷射水平轉()

def 雷射上轉():
    雷射上()
    time.sleep(0.2)
    雷射垂直轉()

def 雷射下轉():
    雷射下()
    time.sleep(0.2)
    雷射垂直轉()


"""
按鈕區
"""
閃阿賓button = tk.Button(master, text="閃阿賓", bg="white", command=閃他,font = 100,width = 18,height=4)
閃阿賓button.grid(row=0, column=0)

吵阿賓button = tk.Button(master, text="吵阿賓",bg="white" , command=吵他,font = 100,width = 18,height=4)
吵阿賓button.grid(row=0, column=1)


振阿賓button = tk.Button(master, text="振阿賓",bg="white" , command=振動他,font = 100,width = 18,height=4)
振阿賓button.grid(row=0, column=2)


餵魷魚絲button = tk.Button(master, text="餵魷魚絲",bg="white" , command=餵魷魚絲,font = 100,width = 18,height=4)
餵魷魚絲button.grid(row=0, column=3)

逗貓棒button = tk.Label(master, text="逗貓棒",bg="white" ,font = 100,width = 18,height=4)
逗貓棒button.grid(row=0, column=4)
#雲台相機
監視雲台右button = tk.Button(master, text="相機向右",bg="white" , command=cam右轉,font = 100,width = 18,height=4)
監視雲台右button.grid(row=4, column=2)

監視雲台左button = tk.Button(master, text="相機向左",bg="white" , command=cam左轉,font = 100,width = 18,height=4)
監視雲台左button.grid(row=4, column=0)


監視雲台上button = tk.Button(master, text="相機向上",bg="white" , command=cam上轉,font = 100,width = 18,height=4)
監視雲台上button.grid(row=3, column=1)

監視雲台下button = tk.Button(master, text="相機向下",bg="white" , command=cam下轉,font = 100,width = 18,height=4)
監視雲台下button.grid(row=5, column=1)

#雲台雷射

雷射筆 = tk.Button(master, text="雷射",bg="white", command=雷射,font = 20,width = 18,height=3)
雷射筆.grid(row=7, column=1)

雷射雲台右button = tk.Button(master, text="雷射向右",bg="white" , command=雷射右轉,font = 100,width = 18,height=4)
雷射雲台右button.grid(row=7, column=2)

雷射雲台右button = tk.Button(master, text="雷射向左",bg="white" , command=雷射左轉,font = 100,width = 18,height=4)
雷射雲台右button.grid(row=7, column=0)


雷射雲台右button = tk.Button(master, text="雷射向上",bg="white" , command=雷射上轉,font = 100,width = 18,height=4)
雷射雲台右button.grid(row=6, column=1)

雷射雲台右button = tk.Button(master, text="雷射向下",bg="white" , command=雷射下轉,font = 100,width = 18,height=4)
雷射雲台右button.grid(row=8, column=1)

"""
標籤
"""
相機 = tk.Label(master, text="相機", fg="white",bg="black",font = 100,width = 18,height=3)
相機.grid(row=4, column=1)




# Exitbutton = tk.Button(master, text="Exit",bg="red", command=master.destroy,width = 20,height=5)
# Exitbutton.grid(row=8, column=0)
master.mainloop()
```
